# 3장 예제

```python
import urllib.request
from bs4 import BeautifulSoup
#기상청 현재 날씨 페이지 읽기
page = urllib.request.urlopen("http://www.weather.go.kr/"+\
                              "weather/observation/currentweather.jsp").read()
soup = BeautifulSoup(page, "html.parser")

#현재 날씨 테이블 검색
table = soup.find("table", {"class":"table_develop3"})

#데이터를 저장할 리스트 생성
data = []
```

```python
#모든 <tr>태그를 검색 각 도시의 데이터를 가져옴
for tr in table.find_all("tr"):
    tds = tr.find_all("td")
    for td in tds:
        if td.find("a"):
            point = td.find("a").text
            temperature = tds[5].text
            humidity = tds[9].text
            data.append([point, temperature, humidity])
            
print(data)

with open("weather.csv","w",encoding="utf8") as file:
    file.write("point, temperature, humidity\n")
    for i in data:
        file.write("{0},{1},{2}\n".format(i[0],i[1],i[2]))
```
