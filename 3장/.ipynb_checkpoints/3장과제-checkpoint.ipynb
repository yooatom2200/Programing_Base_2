# 3장 예제

```python
import urllib.request
from bs4 import BeautifulSoup
#기상청 현재 날씨 페이지 읽기
page = urllib.request.urlopen("http://www.weather.go.kr/"+\
                              "weather/observation/currentweather.jsp").read()
soup = BeautifulSoup(page, "html.parser")

#현재 날씨 테이블 검색
table = soup.find("table", {"class":"table_develop3"})

#데이터를 저장할 리스트 생성
data = []
```

```python
#모든 <tr>태그를 검색 각 도시의 데이터를 가져옴
for tr in table.find_all("tr"):
    tds = tr.find_all("td")
    for td in tds:
        if td.find("a"):
            point = td.find("a").text
            temperature = tds[5].text
            humidity = tds[9].text
            data.append([point, temperature, humidity])
            
print(data)

with open("weather.csv","w",encoding="utf8") as file:
    file.write("point, temperature, humidity\n")
    for i in data:
        file.write("{0},{1},{2}\n".format(i[0],i[1],i[2]))
```

```python
#현재시간에 맞는 날씨정보 찾기
datedata = soup.find('input',{'id':'observation_text'})
print(datedata['value'])

#찾은 데이터 스플릿후 확인, 저장
comma_spl = datedata['value'].split('.')
colon_spl = comma_spl.pop(3).split(':')
data_spl = comma_spl + colon_spl
print(data_spl)
```

```python
#종합
#날씨테이블 데이터 추출
for tr in table.find_all("tr"):
    tds = tr.find_all("td")
    for td in tds:
        if td.find("a"):
            point = td.find("a").text
            temperature = tds[5].text
            humidity = tds[9].text
            data.append([point, temperature, humidity])
            
print(data)

#현재시간에 맞는 날씨정보 찾기
datedata = soup.find('input',{'id':'observation_text'})
print(datedata['value'])

#찾은 데이터 스플릿후 확인, 저장
comma_spl = datedata['value'].split('.')
colon_spl = comma_spl.pop(3).split(':')
data_spl = comma_spl + colon_spl
print(data_spl)

with open("weather.csv","w",encoding="utf8") as file:
    file.write("{0}년 {1}월 {2}일 {3}시 {4}분 현재 날씨\n"\
               .format(data_spl[0],data_spl[1],data_spl[2],data_spl[3],data_spl[4]))
    file.write("point, temperature, humidity\n")
    for i in data:
        file.write("{0},{1},{2}\n".format(i[0],i[1],i[2]))
```

# 임의의 프로그램 작성
---
* 오늘의 운세 불러오기
* 데일리운세(https://unse.daily.co.kr/?p=zodiac)
* 띠와 연도 구분하여 사용자 정보 입력 후 제공

```python
import urllib.request
from bs4 import BeautifulSoup

#페이지 불러오기
ex_page = urllib.request.urlopen("https://unse.daily.co.kr/?p=zodiac").read()
ex_soup = BeautifulSoup(ex_page, "html.parser")
```

```python
#하단 탭 실행전 값 확인을 위한 변수
correct_input = False
try:
    #사용자 정보 입력
    z_year = input("당신이 태어년 해 뒤 두자리를 입력해 주세요! - ")
    zodiac = input("당신의 띠(동물)을 입력해 주세요! - ")

    #띠별 인덱스 구하기
    animals = ['쥐', '소', '호랑이','토끼','용','뱀','말','양','원숭이','닭','개','돼지']
    z_code = "code_" + str(animals.index(zodiac) + 1)
    

    #년도와 띠 체크
    check_y = int(z_year)
    #00년생 이후 12로 나누는게 불가능하여 100추가하여 계산
    if check_y <= 12:
        check_y += 100
    #연도값을 12로 나눈 나머지값이 띠 인덱스와 일치시 통과
    if check_y % 12 != animals.index(zodiac):
        raise Exception#강제 예외처리
    correct_input = True
except:
    print("정확한 값을 입력해 주세요!") 
```

```python
if correct_input == False: 
    print("검색 오류입니다. 띠와 생년을 확인해주세요!")

else:
    #웹페이지 운세목록 추출
    fortune = ex_soup.find("ul", {"id":z_code})

    #사용자 정보 확인
    print("-------------사용자 정보-------------")
    print(z_year + "년생 " + zodiac + "띠의 운세입니다.\n\n")

    #띠 공통 운세
    total_text = fortune.find("p").text
    print("--------------공통 운세--------------")
    print(total_text)

    #띠 연도에 맞는 운세
    print("\n\n-------------연도별 운세-------------")
    for i in fortune.find_all("li"):
        if i.find("span").text == str(z_year) + "년생":
            print(i.find("p").text)
```
